```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import plotly.graph_objects as go
import kaleido
from scipy.integrate import odeint
import plotly.io as pio
import os
import csv
```

```python
df = pd.read_csv('D:/result/res.coconet.dyn.connectR.1.0.blockR.0.5.agentNum.1000.maxTime.500.beta.0.016.e2i.4.5.i2r.8.0.csv')
```

```python
myTime = df.loc[:,'time'].values
S = df.loc[:,'sR'].values
E = df.loc[:,'eR'].values
I = df.loc[:,'iR'].values
R = df.loc[:,'rR'].values
E_con = df.loc[:,'conExpR'].values
R_con = df.loc[:,'conReaR'].values
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=myTime,y=S, mode='lines+markers', name='Susceptible'))
fig.add_trace(go.Scatter(x=myTime,y=E, mode='lines+markers', name='Exposed'))
fig.add_trace(go.Scatter(x=myTime,y=I, mode='lines+markers', name='Infected'))
fig.add_trace(go.Scatter(x=myTime,y=R, mode='lines+markers',name='Recovered'))
fig.add_trace(go.Scatter(x=myTime,y=E_con, mode='lines+markers',name='Expected_contact'))
# fig.add_trace(go.Scatter(x=myTime,y=R_con, mode='lines+markers',name='Real_contact'))


fig.update_layout(title='Simulation of SEIR Model',xaxis_title='Day',
                    yaxis_title='Ratio',title_x=0.5,width=900, height=600)


if not os.path.exists("images"):
    os.mkdir("images")
# fig.write_image("images/seir_simulation.png")
fig.show()
```
